## Project Objectives

1. Perform exploratory data analysis to understand visitor behavior.
2. Identify key factors influencing book purchases.
3. Segment visitors into distinct reader profiles using machine learning.
4. Analyze textual feedback using NLP to extract sentiment and themes.


```python
df.columns = df.columns.str.strip()

for col in df.columns:
    print(f"'{col}'")

```

```python
eda_cols = [
    "Age",
    "Gender",
    "What is your association with Fergusson College?",
    "Which city are you originally from?",
    "What is your profession?",
    "How often do you read books?",
    "Preferred reading format?",
    "Preferred book genres (select multiple)",
    "How many books do you buy per year?",
    "Have you made any/are you planning to make any purchases at the Pune Book Festival?",
    "How has the experience of the book festival been?"
]

eda_df = df[eda_cols].copy()
eda_df.head()
```

```python
eda_df["Age"].value_counts()
```

```python
import matplotlib.pyplot as plt

eda_df["Age"].value_counts().plot(kind="bar")
plt.title("Age Distribution of Visitors")
plt.xlabel("Age Group")
plt.ylabel("Number of Visitors")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
eda_df["How often do you read books?"].value_counts().plot(kind="bar")
plt.title("Reading Frequency of Visitors")
plt.xlabel("Reading Frequency")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
eda_df["Preferred reading format?"].value_counts().plot(kind="pie", autopct="%1.1f%%")
plt.title("Preferred Reading Format")
plt.ylabel("")
plt.show()
```

```python
eda_df["Have you made any/are you planning to make any purchases at the Pune Book Festival?"].value_counts().plot(
    kind="bar"
)
plt.title("Purchase Intent at the Book Festival")
plt.xlabel("Purchase Intent")
plt.ylabel("Count")
plt.show()
```

```python
eda_df["How has the experience of the book festival been?"].plot(kind="hist", bins=10)
plt.title("Distribution of Experience Ratings")
plt.xlabel("Rating")
plt.ylabel("Frequency")
plt.show()
```

```python
purchase_vs_reading = pd.crosstab(
    eda_df["How often do you read books?"],
    eda_df["Have you made any/are you planning to make any purchases at the Pune Book Festival?"],
    normalize="index"
)

purchase_vs_reading
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.heatmap(purchase_vs_reading, annot=True, cmap="Blues", fmt=".2f")
plt.title("Reading Frequency vs Purchase Intent")
plt.ylabel("Reading Frequency")
plt.xlabel("Purchase Intent")
plt.show()
```

```python
age_experience = eda_df.groupby("Age")["How has the experience of the book festival been?"].mean()

age_experience.sort_values().plot(kind="bar")
plt.title("Average Experience Rating by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.show()
```

```python
gender_purchase = pd.crosstab(
    eda_df["Gender"],
    eda_df["Have you made any/are you planning to make any purchases at the Pune Book Festival?"],
    normalize="index"
)

plt.figure(figsize=(6,4))
sns.heatmap(gender_purchase, annot=True, cmap="Greens", fmt=".2f")
plt.title("Gender vs Purchase Intent")
plt.ylabel("Gender")
plt.xlabel("Purchase Intent")
plt.show()
```

```python
format_reading = pd.crosstab(
    eda_df["How often do you read books?"],
    eda_df["Preferred reading format?"],
    normalize="index"
)

plt.figure(figsize=(8,5))
sns.heatmap(format_reading, annot=True, cmap="Oranges", fmt=".2f")
plt.title("Reading Frequency vs Preferred Format")
plt.ylabel("Reading Frequency")
plt.xlabel("Format")
plt.show()
```
